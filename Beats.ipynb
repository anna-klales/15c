## Demonstrating Beats

To use this notebook...

 

*   click in the cell below and press Shift+Enter to run the cell. It will import libraries and define our beats function.
*   Call the function - the third cell has an example of the function call with input frequency values of 1 Hz and 30 Hz. 
*   To run it again with a different choice of frequencies either edit cell 3 or call the function again in a new cell.

```python
# run this cell first

import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

plt.rcParams['font.size'] = '14'

def beats(f1=30,f2=33,tmax=1,Nt=1000,A1=1,A2=1,downloadplot=False):

    tt = np.linspace(0,tmax,1000)


    fig,[ax1,ax2] = plt.subplots(2,1,figsize = (14,6),sharex=True,sharey=True)
    psi1 = A1*np.cos(2*np.pi*f1*tt)
    psi2 = A2*np.cos(2*np.pi*f2*tt)

    ax1.plot(tt,psi1,'r',label='f = '+str(f1)+' Hz',lw = 1)
    ax1.plot(tt,psi2,'b',label='f = '+str(f2)+' Hz',lw = 1)


    psi = psi1 + psi2


    ax2.plot(tt,psi,'k')
    ax2.set_xlabel('Time [s]')
    ax1.set_ylabel('Pressure')
    ax2.set_ylabel('Pressure')

    ax1.legend();
    if downloadplot == True:
        plt.savefig('myplot.pdf',transparent=True)
        files.download('myplot.pdf')
```

```python
#run this cell second

beats(f1=30,f2=1)
```
